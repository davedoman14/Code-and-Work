```python
#Create a function to remove punctuation and non-alphabetic characters#
def remove_punc_filter(word):

    mPattern = re.compile('^[^a-z]+$')
    if (mPattern.match(word)):
        return True
    else:
        return False
```

```python
#Use created function and apply to Paradise Lost#
paradisewords1 = [w for w in paradisewords if not remove_punc_filter(w)]
paradisedictionary['RemovePunctuation'] = len(paradisewords1)
paradisewords1[:50]
```

```python
#Use created function and apply to Leaves of Grass#
leaveswords1 = [w for w in leaveswords if not remove_punc_filter(w)]
leavesdictionary['RemovePunctuation'] = len(leaveswords1)
leaveswords1[:50]
```

```python
#Take a look at original list of stopwords. Will add additional stopwords to this list for each of the writings#
stopwords = nltk.corpus.stopwords.words('english')
stopwords
```

```python
#Examine the tokens and identify those that should be added to the stopwords list#
paradiseaddstopwords = ["'d", "nly", "didst", "'t", "ns", "s_", "th","st", "n","'s","rd","e","thir","thy",
                        "thou","thee","thus","shall","op","ning","nd","cov"]
paradisefinstopwords = stopwords + paradiseaddstopwords
len(paradisefinstopwords)
nostopparadisewords = [w for w in paradisewords1 if not w in paradisefinstopwords]
paradisedictionary['nostopwords'] = len(nostopparadisewords)
```

```python
#Examine the tokens and identify those that should be added to the stopwords list#
leavesaddstopwords = ["s", "d", "walt", "whitman", "i.", "ii.","xvii.","xxi.","xxii.","xxxii.","xxxiii.",
                      "bookxxxv.", "s-self", "st", "lo","'er","thee","ebb","thy","thou","e","en"]
leavesfinstopwords = stopwords + leavesaddstopwords
len(leavesfinstopwords)
nostopleaveswords = [w for w in leaveswords1 if not w in leavesfinstopwords]
leavesdictionary['nostopwords'] = len(nostopleaveswords)
```

```python
#Create Bigram Association Measure#
Bigrammeasures = nltk.collocations.BigramAssocMeasures()
```

```python
#Create the bigram finder for Paradise Lost#
paradiseBCF = BigramCollocationFinder.from_words(nostopparadisewords)

paradiseBCF.apply_freq_filter(4)
paradiseBCFscores = paradiseBCF.score_ngrams(Bigrammeasures.raw_freq)
# check the top 50 scores
for score in paradiseBCFscores[:50]:
    print(score)
```

```python
#Provide results in a format that is better structured#
for score in paradiseBCFscores[:50]:
    print("Words: ", score[0], "\tScore: ", score[1])
```

```python
#Create the bigram finder for Leaves of Grass#
leavesBCF = BigramCollocationFinder.from_words(nostopleaveswords)

leavesBCF.apply_freq_filter(3)
leavesBCFscores = leavesBCF.score_ngrams(Bigrammeasures.raw_freq)
# check the top 50 scores
for score in leavesBCFscores[:50]:
    print(score)
```

```python
#Provide results in a format that is better structured#
for score in leavesBCFscores[:50]:
    print("Words: ", score[0], "\tScore: ", score[1])
```

```python
#Create PMI for Paradise Lost (Bigrams)#
paradisePMIBCF = BigramCollocationFinder.from_words(nostopparadisewords,window_size = 5)

paradisePMIBCF.apply_freq_filter(3)
paradisePMIscores = paradisePMIBCF.score_ngrams(Bigrammeasures.pmi)

for item in paradisePMIscores[:50]:
    print(item)
```

```python
#Create PMI for Leaves of Grass (Bigrams)#
leavesPMIBCF = BigramCollocationFinder.from_words(nostopleaveswords,window_size = 5)

leavesPMIBCF.apply_freq_filter(3)
leavesPMIscores = leavesPMIBCF.score_ngrams(Bigrammeasures.pmi)

for score in leavesPMIscores[:50]:
    print(score)
```

```python
#Create PMI for Paradise Lost (Trigrams)#
paradisePMITCF = TrigramCollocationFinder.from_words(nostopparadisewords, window_size = 5)

paradisePMITCF.apply_freq_filter(3)
paradiesPMITscores = paradisePMITCF.score_ngrams(Trigrammeasures.pmi)
paradiesPMIT = paradisePMITCF.nbest(Trigrammeasures.pmi,50)

for item in paradiesPMITscores[:50]:
    print(item)
```

```python
#Create PMI for Leaves of Grass (Trigrams)#
leavesPMITCF = TrigramCollocationFinder.from_words(nostopleaveswords, window_size = 5)

leavesPMITCF.apply_freq_filter(3)
leavesPMITscores = leavesPMITCF.score_ngrams(Trigrammeasures.pmi)
leavesPMIT = leavesPMITCF.nbest(Trigrammeasures.pmi,50)

for item in leavesPMITscores[:50]:
    print(item)
```
